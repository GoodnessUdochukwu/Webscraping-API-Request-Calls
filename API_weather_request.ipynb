```python
import pandas as pd
import requests
import json
```

```python
def city_weather_data(city_ctry):
    
    weather_data_dict = {'City_Name': [],
                         'Country_Code': [],
                         'City_Population' : [],
                         'Longitude': [],
                         'Latitude': [],
                         'Date_Time': [],
                         'Temp (degC)': [],
                         'Weather': [],
                         'Visibility': [],
                         'Precp_Prob': [],
                         'Wind_Speed(Gust)': []
                        }
    
    for city in city_ctry:
        city_weather_info = requests.get(f'https://api.openweathermap.org/data/2.5/forecast?q={city}&appid=2d0a0a9a66c93f154e24ac10a229d84d&units=metric').json()
        
        Wind_Speed = []
        Wind_Gust = []
        
        for city_weather in city_weather_info['list']:
            weather_data_dict['City_Name'].append(city)
            weather_data_dict['Country_Code'].append(city_weather_info['city']['country'])
            weather_data_dict['City_Population'].append(city_weather_info['city']['population'])
            weather_data_dict['Longitude'].append(city_weather_info['city']['coord']['lon'])
            weather_data_dict['Latitude'].append(city_weather_info['city']['coord']['lat'])
            weather_data_dict['Date_Time'].append(city_weather['dt_txt'])
            weather_data_dict['Temp (degC)'].append(city_weather['main']['temp']) 
            weather_data_dict['Weather'].append(city_weather['weather'][0]['description'])
            weather_data_dict['Visibility'].append(city_weather['visibility'])
            weather_data_dict['Precp_Prob'].append(city_weather['pop'])
            Wind_Speed.append(city_weather['wind']['speed'])
            Wind_Gust.append(city_weather['wind']['gust'])
                
        for a in range(len(Wind_Speed)):
            weather_data_dict['Wind_Speed(Gust)'].append(f'{str(Wind_Speed[a])}({str(Wind_Gust[a])})')
    
    # return dataframe_collection
    return pd.DataFrame(weather_data_dict)
```

```python
full_cities = pd.read_csv('Ngoc_city_data.csv')
full_cities
```

```python
cities_choice = full_cities.loc[0:150,]
cities_choice
```

```python
cities = list(cities_choice.name)
cities
```

```python
weather_data = city_weather_data(cities)
weather_data
```

```python
def get_coords(city_ctry):
    
    city_coord_dict = {'City_Name': [],
                         'Country_Code': [],
                         'Longitude': [],
                         'Latitude': [],
                         }
    
    for city in city_ctry:
        city_weather_info = requests.get(f'https://api.openweathermap.org/data/2.5/forecast?q={city}&appid=2d0a0a9a66c93f154e24ac10a229d84d&units=metric').json()
        
        city_coord_dict['City_Name'].append(city)
        city_coord_dict['Country_Code'].append(city_weather_info['city']['country'])
        city_coord_dict['Longitude'].append(city_weather_info['city']['coord']['lon'])
        city_coord_dict['Latitude'].append(city_weather_info['city']['coord']['lat'])
    
    # return dataframe_collection
    return pd.DataFrame(city_coord_dict)
```
