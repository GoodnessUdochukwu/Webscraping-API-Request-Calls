## **Webscrap Information on Brazil**

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import json
```

```python
url = 'https://en.wikipedia.org/wiki/Federative_units_of_Brazil'
response = requests.get(url)
response.status_code
response.content
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
table_data = soup.select('tbody > tr > td')
table_data = table_data[32:250]
table_data
```

```python
Brazilian_states_demo = {'Name':[],
                         'Code':[],
                         'Capital':[],
                         'Area(km-sq)':[],
                         'Population':[],
                         'Density(per-km-sq)':[],
                         'GDP(R$-mil)':[],
                         'HDI':[]
                        }
                    
for i in range(0, 56, 8):
    Brazilian_states_demo['Name'].append(table_data[i].get_text().strip('\xa0'))
for j in range(65, 185, 8):
    Brazilian_states_demo['Name'].append(table_data[j].get_text().strip('\xa0'))
for k in range(194, 218, 8):
    Brazilian_states_demo['Name'].append(table_data[k].get_text().strip('\xa0'))

Brazilian_states_demo['Name'].append(table_data[56].get_text().strip('\xa0'))
Brazilian_states_demo['Name'].append(table_data[185].get_text().strip('\xa0'))


for a in range(1, 57, 8):
    Brazilian_states_demo['Code'].append(table_data[a].get_text())
for b in range(66, 186, 8):
    Brazilian_states_demo['Code'].append(table_data[b].get_text())
for c in range(195, 218, 8):
    Brazilian_states_demo['Code'].append(table_data[c].get_text())
    
Brazilian_states_demo['Code'].append(table_data[57].get_text())
Brazilian_states_demo['Code'].append(table_data[186].get_text())


for a in range(2, 58, 8):
    Brazilian_states_demo['Capital'].append(table_data[a].get_text())
for b in range(67, 187, 8):
    Brazilian_states_demo['Capital'].append(table_data[b].get_text())
for c in range(196, 218, 8):
    Brazilian_states_demo['Capital'].append(table_data[c].get_text())
    
Brazilian_states_demo['Capital'].append(table_data[58].get_text())
Brazilian_states_demo['Capital'].append(table_data[187].get_text())  


for a in range(3, 59, 8):
    Brazilian_states_demo['Area(km-sq)'].append(table_data[a].get_text())
for b in range(68, 188, 8):
    Brazilian_states_demo['Area(km-sq)'].append(table_data[b].get_text())
for c in range(197, 218, 8):
    Brazilian_states_demo['Area(km-sq)'].append(table_data[c].get_text())
    
Brazilian_states_demo['Area(km-sq)'].append(table_data[60].get_text())
Brazilian_states_demo['Area(km-sq)'].append(table_data[189].get_text()) 


for a in range(4, 60, 8):
    Brazilian_states_demo['Population'].append(table_data[a].get_text())
for b in range(69, 189, 8):
    Brazilian_states_demo['Population'].append(table_data[b].get_text())
for c in range(198, 218, 8):
    Brazilian_states_demo['Population'].append(table_data[c].get_text())
    
Brazilian_states_demo['Population'].append(table_data[61].get_text())
Brazilian_states_demo['Population'].append(table_data[190].get_text()) 


for a in range(5, 61, 8):
    Brazilian_states_demo['Density(per-km-sq)'].append(table_data[a].get_text())
for b in range(70, 190, 8):
    Brazilian_states_demo['Density(per-km-sq)'].append(table_data[b].get_text())
for c in range(199, 218, 8):
    Brazilian_states_demo['Density(per-km-sq)'].append(table_data[c].get_text())
    
Brazilian_states_demo['Density(per-km-sq)'].append(table_data[62].get_text())
Brazilian_states_demo['Density(per-km-sq)'].append(table_data[191].get_text()) 


for a in range(6, 62, 8):
    Brazilian_states_demo['GDP(R$-mil)'].append(table_data[a].get_text())
for b in range(71, 191, 8):
    Brazilian_states_demo['GDP(R$-mil)'].append(table_data[b].get_text())
for c in range(200, 218, 8):
    Brazilian_states_demo['GDP(R$-mil)'].append(table_data[c].get_text())
    
Brazilian_states_demo['GDP(R$-mil)'].append(table_data[63].get_text())
Brazilian_states_demo['GDP(R$-mil)'].append(table_data[192].get_text()) 


for a in range(7, 63, 8):
    Brazilian_states_demo['HDI'].append(table_data[a].get_text().strip('\n'))
for b in range(72, 192, 8):
    Brazilian_states_demo['HDI'].append(table_data[b].get_text().strip('\n'))
for c in range(201, 218, 8):
    Brazilian_states_demo['HDI'].append(table_data[c].get_text().strip('\n'))
    
Brazilian_states_demo['HDI'].append(table_data[64].get_text().strip('\n'))
Brazilian_states_demo['HDI'].append(table_data[193].get_text().strip('\n')) 


Brazilian_states_demo
```
